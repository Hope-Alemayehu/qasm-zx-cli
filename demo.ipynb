```python
import pyzx as zx
from pyzx import Circuit, simplify
import numpy as np
import matplotlib.pyplot as plt
```

```python
qasm = """
OPENQASM 2.0;
include "qelib1.inc";

qreg q[4];

h q[0];
x q[0];
x q[1];

cx q[0], q[2];
cx q[1], q[2];
cx q[0], q[3];
"""
circuit = Circuit.from_qasm(qasm)
```

```python
#option 2 to upload qasm file directly
# circuit = Circuit.from_qasm_file("example.qasm")
```

```python
g = circuit.to_graph()
basic_g = g
```

```python

zx.draw_matplotlib(basic_g, figsize=(10,2), h_edge_draw='box')
```
